```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
#importation des bib 

# Importation de la base de donné
dataset = pd.read_csv('data.csv')
x = dataset.iloc[:, :-1].values 
# x prend la premiere colone 
y = dataset.iloc[:, -1].values 
# y prend la 2 eme colone 


from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 1/3, random_state = 0) 
#2/3 pour l'apprentissage et 1/3 pour le test 


from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
#on fait appel à la commande de la régression et on cherches les meilleurs résultats possibles


y_pred = regressor.predict(x_test)
#prediction de y selon x utilisé dans la phase test



plt.scatter(x_train, y_train, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Heights vs Weights (Training set)')
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.show()
#shéma de la phase apprentissage

# Visualising the Test set results
plt.scatter(x_test, y_test, color = 'red')
plt.plot(x_train, regressor.predict(x_train), color = 'blue')
plt.title('Heights vs Weights (Test set)')
plt.xlabel('Heights')
plt.ylabel('Weights')
plt.show()
#shéma de la phase test



y_pred2= regressor.predict([[1.90]])
print(y_pred2)
#prediction de y selon x=1.90
print("----------------------")
print('Train Score: ', regressor.score(x_train, y_train)) 
print('Test Score: ', regressor.score(x_test, y_test)) 
#calcul de taux de reconnaissance 
```
